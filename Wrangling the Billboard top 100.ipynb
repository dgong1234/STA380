# Wrangling The Billboard Top 100

```python
from pandas import Series, DataFrame
import pandas as pd
%pylab inline
```

## Part a.

```python
bb100 = pd.read_csv('data/billboard.csv')
```

```python
bb100w = bb100.groupby(['song', 'performer']).size().reset_index(name='count')
bb100w = bb100w.sort_values(by='count', ascending=False)[:10]
bb100w
```

This table shows the 10 most popular songs since 1958, measured by the weeks on the top chart. As you can see, the most popular song is Radioactive with 87 weeks on the chart, and the least popular of the 10 is Before He Cheats, with 64 weeks on the chart.

## Part b.

```python
bb100y = bb100.groupby(['year','song', 'performer']).size().reset_index(name='count')
bb100y = bb100y.sort_values(by='year')
bb100yu = bb100y.groupby(['year']).size().reset_index(name = 'num songs')
bb100yu
```

```python
plot(bb100yu['year'], bb100yu['num songs'], marker='None', linestyle='-')
xlabel('Year')
ylabel('Number of songs')
show()
```

It seems that the musical diversity is changing over time. it starts quite low at the start in 1958 and seems to peak at 850 around 1967 or so. It then goes on a steady decline where it hits a low at around 2002 and then it starts climbing back up and peaks at around 2019, where it then drops back to a low in 2021.

## Part c.

```python
bb100a = bb100.groupby(['song', 'performer']).size().reset_index(name='count')
bb100a = bb100a.sort_values(by='count', ascending=False)
bb100a10 = bb100a[bb100a['count']>= 10]
bb100a10
```

```python
bb10019 = bb100a10.groupby(['performer']).size().reset_index(name='count')
bb10019 = bb10019[bb10019['count']>= 30]
bb10019
```

```python
bb10019.plot(x= 'performer', y = 'count',kind='bar')
xlabel('Performer')
ylabel('Number of songs')
```

This bar plot shows the number of songs that each performer had over their career that were at least 10 week hits from 1958 - 2021. The 19 performers selected were all performers that had at least 30 songs that were 10 week hits. It shows that Elton John had the most at 52 hits, while the lowest was Brad Paisley, Daryl Hall John Oates, George Strait, Jason Aldean, and Neil Diamond who were all at 30 hits. Although these are the lowest, it is still very impressive as they are still top 19 artist in terms of number of hits ver 30 that were 10 week hits.
